# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending 645-675 per student actually underperformed compared to schools with smaller budgets (585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Import dependencies and setup
import pandas as pd

# Load the csv data
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data files and store into Pandas DataFrames
# School Data has School ID, school_name, type, size, and budget
# Student Data has Student ID, student_name, gender, grade, school_name, reading_score, and math_score
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset  
# Left join in order to get all of the student data and matching school data; data is joined using school name which both data files have
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
# We can use the school_data file instead of the combined file to avoid dupes
school_count = school_data.count()["school_name"]
school_count
```

```python
# Calculate the total number of students
# We can use the student_data file instead of the merged file to avoid potential duplicates
student_count = student_data["Student ID"].count()
student_count
```

```python
# Calculate the total budget
# We can use the school_data file instead of the combined file to avoid double counting dupes
total_budget = school_data["budget"].sum()
total_budget
```

```python
# Calculate the average (mean) math score
# We can use either student_data or school_data_complete data frames for this calculation
average_math_score = student_data["math_score"].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
# We can use either student_data or school_data_complete data frames for this calculation
average_reading_score = student_data["reading_score"].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
# We can use either student_data or school_data_complete data frames for this calculation
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / (student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passeed reading (hint: look at how the math percentage was calculated) 
# We can use either student_data or school_data_complete data frames for this calculation 
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / (student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
# We can use either student_data or school_data_complete data frames for this calculation
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  (student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({"Total No. of Schools": [school_count], 
                                "Total Students": [student_count],
                                "Total Budget": [total_budget],
                                "Average Math Score": [average_math_score],
                                "Average Reading Score": [average_reading_score],
                                "Students Passed Math %": [passing_math_percentage],
                                "Students Passed Reading %": [passing_reading_percentage],
                                "Students Passed Both Math & Reading %": [overall_passing_rate]})

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:,.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:,.2f}".format)
district_summary["Students Passed Math %"] = district_summary["Students Passed Math %"].map("{:,.2f}%".format)
district_summary["Students Passed Reading %"] = district_summary["Students Passed Reading %"].map("{:,.2f}%".format)
district_summary["Students Passed Both Math & Reading %"] = district_summary["Students Passed Both Math & Reading %"].map("{:,.2f}%".format)

# Display the DataFrame
district_summary
```

## School Summary

```python
# Get the school types
school_types = school_data.set_index(["school_name"])["type"]
school_types
```

```python
# Calculate the total student count by school name
# We will need to use the merge data school_data_complete to calculate the below
school_data_group = school_data_complete.groupby(["school_name"])
per_school_counts = school_data_group["Student ID"].count()
per_school_counts
```

```python
# Calculate the total school budget 
per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
# Budget per student
per_student_budget = per_school_budget / per_school_counts
```

```python
# Calculate the average test scores
avg_school_math = school_data_complete.groupby(["school_name"]).mean()["math_score"]
avg_school_reading = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
```

```python
# Calculate the number of schools with math scores of 70 or higher
school_passing_math = school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name")["Student ID"].count()
school_passing_math
```

```python
# Calculate the number of schools with reading scores of 70 or higher
school_passing_reading = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name")["Student ID"].count()
school_passing_reading
```

```python
# Use the provided code to calculate the schools that passed both math and reading with scores of 70 or higher
passing_math_and_reading = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)].groupby("school_name")["Student ID"].count()
passing_math_and_reading
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = school_passing_math / (per_school_counts) * 100
per_school_passing_reading = school_passing_reading / (per_school_counts) * 100
overall_passing_rate = passing_math_and_reading / (per_school_counts) * 100
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame({"School Type": school_types,
                                    "Total Students": per_school_counts,
                                    "Total School Budget": per_school_budget,
                                    "Per Student Budget": per_student_budget,
                                    "Average Math Score": avg_school_math,
                                    "Average Reading Score": avg_school_reading,
                                    "Students Passed Math %": per_school_passing_math,
                                    "Students Passed Reading %": per_school_passing_reading,
                                    "Overall Passing %": overall_passing_rate })

# Only formatting School Budget and Per Student Budget because the rest will be used for calculation and cannot be a string type
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
# Commented out the formatting code for the rest 
#per_school_summary["Average Math Score"] = per_school_summary["Average Math Score"].map("{:,.2f}".format)
#per_school_summary["Average Reading Score"] = per_school_summary["Average Reading Score"].map("{:,.2f}".format)
#per_school_summary["Students Passed Math %"] = per_school_summary["Students Passed Math %"].map("{:,.2f}%".format)
#per_school_summary["Students Passed Reading %"] = per_school_summary["Students Passed Reading %"].map("{:,.2f}%".format)
#per_school_summary["Overall Passing %"] = per_school_summary["Overall Passing %"].map("{:,.2f}%".format)

# Display the DataFrame
per_school_summary
```

## Scores by School Spending

```python
# For this section, we will create a table that breaks down school performance based on average spending ranges (per student).

# We will use the per_school_summary data frame created earlier since it has the Per Student Budget
# Print the the column names to use in the code
print(per_school_summary.columns)
# Check data type of Per Student Budget to confirm we can use the values for the bins
print(per_school_summary.dtypes["Per Student Budget"])
# Check the max & min average spending ranges (per student) to confirm the range of the bins are correct
print(per_school_summary["Per Student Budget"].max())
print(per_school_summary["Per Student Budget"].min())


# Establish the bins as stated in the homework instruction 
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]

```

```python
# Before copying, might have to add a code to convert the Per Student Budget from string to integer IF I include formatting when creating the School Summary data frame
# For now, I just commented out the code to format the data
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].str.replace("$", "").astype('float')
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(per_school_summary["Per Student Budget"], spending_bins, labels=spending_labels, include_lowest=True)
school_spending_df.head()
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby("Spending Ranges (Per Student)").mean()["Average Math Score"]
spending_reading_scores = school_spending_df.groupby("Spending Ranges (Per Student)").mean()["Average Math Score"]
spending_passing_math = school_spending_df.groupby("Spending Ranges (Per Student)").mean()["Students Passed Math %"]
spending_passing_reading = school_spending_df.groupby("Spending Ranges (Per Student)").mean()["Students Passed Reading %"]
overall_passing_spending = school_spending_df.groupby("Spending Ranges (Per Student)").mean()["Overall Passing %"]
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({"Average Math Score": spending_math_scores,
                                "Average Reading Score": spending_reading_scores,
                                "Students Passed Math %": spending_passing_math,
                                "Students Passed Reading %": spending_passing_reading,
                                "Overall Passing %": overall_passing_spending})

# Display results
spending_summary
```
